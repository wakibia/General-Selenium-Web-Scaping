```python
## import library
from selenium import webdriver
import pandas as pd
import numpy as np
import time
# handle cookies
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
from selenium.webdriver.support import expected_conditions as EC
```

>- The website `https://www.buyrentkenya.com` can be divided into two, there are the items that have been listed in the buy and the rent 
>- We are interested in scraping the following elements
1. title of the listing (it's like a description of the size of the listed property)
2. price of the property
3. number of bedrooms
4. number of bathrooms
5. size of the property(in sq feets)
6. Location of the property
7. Real Estate Agent Name under which the property has been listed

```python
## website to scrape

buykenyawebsite = "https://www.buyrentkenya.com/houses-for-sale"
path = 'C:/Users/cmwak/chrome_driver/chromedriver'

## define a driver variable'
driver = webdriver.Chrome(path)

## open the driver
driver.get(buykenyawebsite)

## maximize the window to display all over the screen
driver.maximize_window()




## create the list to store the data
property_name = []
property_price = []
property_location = []
property_bedrooms = []
property_bathrooms = []
property_size = []
real_estate_agent = []


# container

container = driver.find_element_by_xpath('//div[contains(@data-cy, "listing-section")]')
propertys = container.find_element_by_xpath('.//div[contains(@data-bi,"product" )]')

## accept cookies
try:
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//button[contains(@class, "cookie")]'))).click()
except:
    pass

## get data from the first container
## property_name = propertys.find_element_by_xpath('.//span[contains(@class, "md:inline")]').text
#property_price = container.find_element_by_css_selector('a[href="https://www.example.com/link"]').text
## property_location = propertys.find_element_by_xpath('.//p[contains(@class, "text-sm")]').text
## property_bedrooms = propertys.find_element_by_xpath('.//span[contains(@data-cy, "card-beds")]').text
## property_bathrooms = propertys.find_element_by_xpath('.//div[contains(@data-cy, "card-baths")]').text
## property_size = container.find_element_by_xpath('.//span[contains(@data-cy, "card-area")]').text
## real_estate_agent = propertys.find_element_by_xpath('.//img[contains(@src, "agency-thumb")]').get_attribute('alt')


# scrape the data from entire page
properties = container.find_elements_by_xpath('.//div[contains(@data-bi,"product" )]')
for listing in properties:
    ## get data from the first container
    try:
        name = listing.find_element_by_xpath('.//span[contains(@class, "md:inline")]').text
        property_name.append(name)
    except:
        property_name.append(np.nan)
    try:
        location = listing.find_element_by_xpath('.//p[contains(@class, "text-sm")]').text
        property_location.append(location)
    except:
        property_location.append(np.nan)
    try:
        bedrooms = listing.find_element_by_xpath('.//span[contains(@data-cy, "card-beds")]').text
        property_bedrooms.append(bedrooms)
    except:
        property_bedrooms.append(np.nan)
    try:
        bathrooms = listing.find_element_by_xpath('.//div[contains(@data-cy, "card-baths")]').text
        property_bathrooms.append(bathrooms)
    except:
        property_bathrooms.append(np.nan)
    try:
        agent = listing.find_element_by_xpath('.//img[contains(@src, "agency-thumb")]').get_attribute('alt')
        real_estate_agent.append(agent)
    except:
        real_estate_agent.append(np.nan)
    try:
        price = listing.find_element_by_xpath('.//p[contains(@class, "font-bold text-xl leading-7 text-grey-900")]').text
        property_price.append(price)
    except:
        property_price.append(np.nan)
    try: 
        size = listing.find_element_by_xpath('.//span[contains(@data-cy, "card-area")]').text
        property_size.append(size) 
    except:
        property_size.append(np.nan)



```

```python
# create a data frame
property_df = pd.DataFrame({'property_name': property_name,
                            'property_price': property_price,
                            'property_location': property_location,
                            'property_bedrooms': property_bedrooms,
                            'property_bathrooms': property_bathrooms,
                            'property_size': property_size,
                            'real_estate_agent': real_estate_agent
                            })
property_df.to_csv('property_df.csv', index=False)
property_df.head(10)
```

```python
## website to scrape

buykenyawebsite = "https://www.buyrentkenya.com/houses-for-sale"
path = 'C:/Users/cmwak/chrome_driver/chromedriver'


## get all the data now
driver = webdriver.Chrome(path)

## open the driver
driver.get(buykenyawebsite)

## maximize the window to display all over the screen
driver.maximize_window()

pagination = driver.find_element_by_xpath('//ul[contains(@class, "pagination-page-nav")]')

## get the number of pages
pages = pagination.find_elements_by_xpath('.//li[contains(@class, "page-item")]')

## get the last page
last_page = int(pages[-1].text)


# container
container = driver.find_element_by_xpath('//div[contains(@data-bi, "product")]')

## accept cookies
try:
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//button[contains(@class, "cookie")]'))).click()
except:
    pass


current_page = 1

## create the list to store the data
property_name = []
property_price = []
property_location = []
property_bedrooms = []
property_bathrooms = []
property_size = []
real_estate_agent = []

## scrape only 30 pages for convienience sake
while current_page <= last_page:
   ## time.sleep(2) ## wait for the page to load for 2 seconds [implicit wait]
    ## get the elements
    ## maximum waiting time is 5 seconds
    container = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//div[contains(@data-cy, "listing-section")]')))
    
    #container = driver.find_element_by_class_name("adbl-impression-container")
    
    properties = WebDriverWait(container, 5).until(EC.presence_of_all_elements_located((By.XPATH, './/div[contains(@data-bi,"product" )]')))

    #products = container.find_elements_by_xpath('.//li[contains(@class, "productListItem")]')
    for listing in properties:

        ## get data from the first container
        try:
            name = listing.find_element_by_xpath('.//span[contains(@class, "md:inline")]').text
            property_name.append(name)
        except:
            property_name.append(np.nan)
        try:
            location = listing.find_element_by_xpath('.//p[contains(@class, "text-sm")]').text
            property_location.append(location)
        except:
            property_location.append(np.nan)
        try:
            bedrooms = listing.find_element_by_xpath('.//span[contains(@data-cy, "card-beds")]').text
            property_bedrooms.append(bedrooms)
        except:
            property_bedrooms.append(np.nan)
        try:
            bathrooms = listing.find_element_by_xpath('.//div[contains(@data-cy, "card-baths")]').text
            property_bathrooms.append(bathrooms)
        except:
            property_bathrooms.append(np.nan)
        try:
            agent = listing.find_element_by_xpath('.//img[contains(@src, "agency-thumb")]').get_attribute('alt')
            real_estate_agent.append(agent)
        except:
            real_estate_agent.append(np.nan)
        try:
            price = listing.find_element_by_xpath('.//p[contains(@class, "font-bold text-xl leading-7 text-grey-900")]').text
            property_price.append(price)
        except:
            property_price.append(np.nan)
        try: 
            size = listing.find_element_by_xpath('.//span[contains(@data-cy, "card-area")]').text
            property_size.append(size) 
        except:
            property_size.append(np.nan)

    current_page += 1
    try:
        # navigate to the last page
        # select the last link

        next_page_link = driver.find_element_by_xpath('//a[contains(@class ,"w-32 p-3")]')
        next_page_link.click()
    except:
        pass



driver.quit()


```

```python
# create a df
properties_df_ext = pd.DataFrame({'property_name': property_name,
                                'property_price': property_price,
                                'property_location': property_location,
                                'property_bedrooms': property_bedrooms,
                                'property_bathrooms': property_bathrooms,
                                'property_size': property_size,
                                'real_estate_agent': real_estate_agent
                                })

properties_df_ext.to_csv('properties.csv', index=False)    
```
