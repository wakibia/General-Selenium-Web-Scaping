## get the links from a page

```python
## import modules
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
```

```python
## url 
url = 'https://subslikescript.com/movies'

## url contents
response = requests.get(url=url)

## create the soup object
soup_obj = BeautifulSoup(response.text, 'lxml')

## movies are contained in this unordered list
mvs = soup_obj.find('article', class_="main-article")
```

```python
for pg in range(1, 100):
    url_all = 'https://subslikescript.com/movies?page='
    pg_url = url_all + str(pg)

    response = requests.get(pg_url)

    ## create the soup object
    soup_obj = BeautifulSoup(response.text, 'lxml')

    ## movies are contained in this unordered list
    mvs = soup_obj.find_all('article', class_="main-article")

    mvs_links = []
    mvs_full_link = []
    mv_titles = []

    for mv in mvs:

        mvs_link = mv.find_all('a', href=True)
        titles = mv.find_all('li')

        for link in mvs_link:
            mvs_links.append(link['href'])    
        for x in titles:
            mv_titles.append(str(x))    

```

```python
## get the url links
mvs_link = mvs.find_all('a', href=True)
mvs_links = []
for link in mvs_link:
    mvs_links.append(link['href'])
```

```python
mvs_links
```

```python
## get the movie link
## here is the root link
mvs_root = 'https://subslikescript.com/'
mvs_full_link = []
for mvs_end in mvs_links:
    ##mvs_full = re.sub(r'[-\d*]', '', mvs_end)
    mvs_full_link.append(mvs_root+mvs_end)

mvs_full_link
```

```python
## get the movie title
titles = mvs.find_all('li')
```

```python
## convert the result set to a list of string
mv_titles = []
for x in titles:
    mv_titles.append(str(x))
```

```python
mv_titles
```

```python
## extract year
year = [re.findall(r'\d+', x) for x in mv_titles]
year = flatten(year)
## remove all the numbers that are not 4
movie_yr = flatten([re.findall(r'\d{4}$', x) for x in year])
```

```python
movie_titles = [re.sub(r'(<li>)|(<.li>)', '', x) for x in mv_titles]
```

```python
## data frame for the movie titles and the links
mv_dict = {'movie_title': movie_titles, 'link': mvs_full_link, 'movir_year': movie_yr}
movie_df = pd.DataFrame(mv_dict)
movie_df
```
