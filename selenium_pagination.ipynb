```python
## import library
from selenium import webdriver
import pandas as pd
import numpy as np
```

## <center><b>Handling Pagination with Selenium</b></center>

>- begin with the first page

### What to scrape

>- Title
>- Length
>- Release date
>- Language
>- Author


```python
## website to scrape

website = "https://www.audible.com/search"
path = 'C:/Users/cmwak/chrome_driver/chromedriver'

## define a driver variable'
driver = webdriver.Chrome(path)

## open the driver
driver.get(website)

## maximize the window to display all over the screen
driver.maximize_window()

## get the elements
container = driver.find_element_by_class_name("adbl-impression-container")


products = container.find_elements_by_xpath('.//li[contains(@class, "productListItem")]')

if products and len(products):
    print(f'Found {len(products)}')
else:
    print("No Links found")



titles = []
authors = []
runtimes = []

for product in products:
    ## the dot here indicates that the current context is product
    titles.append(product.find_element_by_xpath('.//h3[contains(@class,"bc-heading")]').text)
    authors.append(product.find_element_by_xpath('.//li[contains(@class,"authorLabel")]').text)
    runtimes.append(product.find_element_by_xpath('.//li[contains(@class,"runtimeLabel")]').text)



driver.quit()
```

```python
## create a dataframe
audio_books = pd.DataFrame(
    {
    "title" : titles, "author": authors, "length": runtimes
    }
)
audio_books.to_csv('Audio_Books.csv', index=False)
```
