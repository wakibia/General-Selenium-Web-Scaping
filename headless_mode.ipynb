## <center>Headless Mode</center>

>- Is an option that selenium has that allows us to run the bot we create on the background
>- This is important when you have other tasks and you don't want to see browsers opening over and over again



```python
## import library
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import pandas as pd
import numpy as np
```

```python
##set options
options = Options()
## activate headless options
options.headless = True
options.add_argument('window-size=1920x1080')

## The bigger the screen the more data you  get


## website to scrape

website = "https://www.audible.com/search"
path = 'C:/Users/cmwak/chrome_driver/chromedriver'

## define a driver variable'
driver = webdriver.Chrome(service=Service(path), options=options)

## open the driver
driver.get(website)

## maximize the window to display all over the screen
driver.maximize_window() ## this only works when headless modde is on

## get the elements
container = driver.find_element(By.CLASS_NAME, "adbl-impression-container")


products = container.find_elements(By.XPATH, './/li[contains(@class, "productListItem")]')

if products and len(products):
    print(f'Found {len(products)}')
else:
    print("No Links found")



titles = []
authors = []
runtimes = []

for product in products:
    ## the dot here indicates that the current context is product
    book_title = product.find_element(By.XPATH, './/h3[contains(@class,"bc-heading")]').text
    titles.append(book_title)
    ## print(book_title)
    authors.append(product.find_element(By.XPATH, './/li[contains(@class,"authorLabel")]').text)
    runtimes.append(product.find_element(By.XPATH, './/li[contains(@class,"runtimeLabel")]').text)



driver.quit()
```

```python
## create a dataframe
audio_books = pd.DataFrame(
    {
    "title" : titles, "author": authors, "length": runtimes
    }
)
audio_books.to_csv('Audio_Books_headless_mode.csv', index=False)
```
